```python
import pandas as pd
import numpy as np
```

```python
data  = pd.read_csv('./data/Exam_Score_Prediction.csv')
print(data.head())
```

```python
data = data.drop(columns=['student_id'])
data = data.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

print(data['exam_difficulty'].unique())
print(data['internet_access'].unique())
print(data['gender'].unique())
print(data['sleep_quality'].unique())
print(data['facility_rating'].unique())
```

```python
label1 = ['easy', 'moderate', 'hard']
lebel2 = ['no', 'yes']
label3 = ['poor', 'average', 'good']

oe = OrdinalEncoder(categories=[label1, lebel2, label3])
data[['exam_difficulty', 'internet_access', 'sleep_quality']] = oe.fit_transform(data[['exam_difficulty', 'internet_access', 'sleep_quality']])
data.head()
```

```python
oe2 = OrdinalEncoder(categories=[['low', 'medium', 'high']])
data[['facility_rating']] = oe2.fit_transform(data[['facility_rating']])
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop = 'first', sparse_output=False)
#perform one hot encoding on course and study_method
encoded_features = ohe.fit_transform(data[['course', 'study_method']])
encoded_df = pd.DataFrame(encoded_features, columns=ohe.get_feature_names_out(['course', 'study_method']))
data = pd.concat([data.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)
data = data.drop(columns=['course', 'study_method'])
data.head()
```

```python
#Encode the gender
le = LabelEncoder()
data['gender'] = le.fit_transform(data['gender'])
data.head()
```

```python
# Create target variable by binning exam_score split into F, D, C, B, A, S

labels = ['F', 'D', 'C', 'B', 'A', 'S']
data['grade'] = pd.cut(data['exam_score'], bins=6, labels = labels, right=False)
data = data.drop(columns=['exam_score'])
data.head()
```

```python
oe = OrdinalEncoder(categories=[labels])
data[['grade']] = oe.fit_transform(data[['grade']])
data.head()
```

```python
data.to_csv('./data/Exam_Score_Prediction_Cleaned.csv', index=False)
```
