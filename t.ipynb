```python
import pandas as pd
import numpy as np
```

```python
data  = pd.read_csv('./data/fruit_classification_dataset.csv')
data.head()
```

```python
print(data['shape'].value_counts())
print(data['color'].value_counts())
print(data['taste'].value_counts())
print(data['fruit_name'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['shape'] = le.fit_transform(data['shape'])
data['color'] = le.fit_transform(data['color'])
data['taste'] = le.fit_transform(data['taste'])
data['fruit_name'] = le.fit_transform(data['fruit_name'])

data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_to_scale = ['size (cm)', 'weight (g)', 'avg_price (₹)']
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])
data.head()
```

```python
data.to_csv('./data/fruit_classification_processed.csv', index=False)
```
